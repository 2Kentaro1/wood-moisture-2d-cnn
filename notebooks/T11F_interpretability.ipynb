```python
from pathlib import Path
import os
import sys

# Optional: uncomment and edit when the repository is in a custom Colab/Drive path.
# os.environ['PROJECT_ROOT'] = '/content/drive/MyDrive/NIR_CNN_2D'

def find_project_root():
    override = os.environ.get('PROJECT_ROOT', '').strip()
    if override:
        root = Path(override).expanduser().resolve()
        if (root / 'src' / 'training' / 'train_regression.py').exists():
            return root
        raise FileNotFoundError(f'PROJECT_ROOT was set, but src was not found: {root}')
    starts = [Path.cwd(), *Path.cwd().parents, Path('/content'), Path('/content/drive/MyDrive'), Path('/workspace'), Path('/mnt/data')]
    for start in starts:
        if (start / 'src' / 'training' / 'train_regression.py').exists():
            return start
    for base in [Path('/content'), Path('/content/drive/MyDrive'), Path('/workspace'), Path('/mnt/data')]:
        if base.exists():
            for hit in base.rglob('train_regression.py'):
                if hit.as_posix().endswith('src/training/train_regression.py'):
                    return hit.parents[2]
    raise FileNotFoundError('Project root not found. Clone/copy the full repository to Colab, or set os.environ["PROJECT_ROOT"].')

PROJECT_ROOT = find_project_root()
os.chdir(PROJECT_ROOT)
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))
print(f'PROJECT_ROOT={PROJECT_ROOT}')
if not (PROJECT_ROOT / 'data' / 'train.csv').exists():
    print('WARNING: data/train.csv was not found under PROJECT_ROOT.')
if 'OUTPUT_DIR' not in os.environ and Path('/content/drive/MyDrive').exists():
    os.environ['OUTPUT_DIR'] = '/content/drive/MyDrive/wood-moisture-2d-cnn-outputs'
Path(os.environ.get('OUTPUT_DIR', 'outputs')).mkdir(parents=True, exist_ok=True)
print(f"OUTPUT_DIR={os.environ.get('OUTPUT_DIR', 'outputs')}")
```

```python
from pathlib import Path
import numpy as np
import torch
from src.data.load_data import build_multiview_input, load_train_test
from src.training.trainer import load_model_checkpoint
from src.interpret.saliency import gradient_saliency
from src.interpret.attention_maps import batch_mean_importance
from src.interpret.visualization import plot_view_wavelength_heatmap

output_dir = Path(os.environ.get('OUTPUT_DIR', 'outputs'))
heatmap_path = output_dir / 'heatmaps' / 'mc_saliency_fold0.png'
importance_path = output_dir / 'heatmaps' / 'mc_importance.npy'
if heatmap_path.exists() and importance_path.exists():
    print('skip: interpretability outputs already exist')
else:
    train_frame, _ = load_train_test('.')
    x = build_multiview_input(train_frame)
    model, ckpt = load_model_checkpoint(output_dir / 'models' / 'mc_fold0.pt')
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model = model.to(device)
    xb = torch.tensor((x[:32] - ckpt['mean']) / ckpt['std'], dtype=torch.float32, device=device)
    sal = gradient_saliency(model, xb, ckpt['task_type'])
    importance = batch_mean_importance(sal)
    importance_path.parent.mkdir(parents=True, exist_ok=True)
    np.save(importance_path, importance)
    plot_view_wavelength_heatmap(importance, train_frame.wavelengths, 'MC saliency fold0', heatmap_path)
```
