# T00 Colab Setup

Clone or update the GitHub repository, mount Google Drive, install dependencies, and set persistent output paths for the training notebooks.


```python
from getpass import getpass
from pathlib import Path
import os

GITHUB_REPO = '2Kentaro1/wood-moisture-2d-cnn'
PROJECT_DIR = Path('/content/wood-moisture-2d-cnn')
DRIVE_OUTPUT_DIR = Path('/content/drive/MyDrive/wood-moisture-2d-cnn-outputs')

from google.colab import drive
drive.mount('/content/drive')
DRIVE_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

# For a public repository, leave token empty.
# For a private repository, paste a GitHub fine-grained token with repository read access.
token = getpass('GitHub token (empty for public repo): ')
repo_url = f'https://github.com/{GITHUB_REPO}.git'
clone_url = repo_url if not token else f'https://{token}@github.com/{GITHUB_REPO}.git'

if PROJECT_DIR.exists():
    %cd {PROJECT_DIR}
    !git pull
else:
    %cd /content
    !git clone {clone_url} {PROJECT_DIR}
    %cd {PROJECT_DIR}

os.environ['PROJECT_ROOT'] = str(PROJECT_DIR)
os.environ['OUTPUT_DIR'] = str(DRIVE_OUTPUT_DIR)
print('PROJECT_ROOT=', os.environ['PROJECT_ROOT'])
print('OUTPUT_DIR=', os.environ['OUTPUT_DIR'])
print('src exists:', (PROJECT_DIR / 'src').exists())
print('train exists:', (PROJECT_DIR / 'data' / 'train.csv').exists())
print('test exists:', (PROJECT_DIR / 'data' / 'test.csv').exists())
print('drive output exists:', DRIVE_OUTPUT_DIR.exists())
```
